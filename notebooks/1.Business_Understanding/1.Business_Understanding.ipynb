# Business Understanding (ENGLISH BELLOW)

## 1. Visão Geral do Projeto

Este projeto tem como objetivo analisar a deterioração significativa do risco de crédito (*Significant Increase in Credit Risk – SICR*) no contexto do IFRS 9, utilizando uma abordagem Top-Down aplicada aos dados agregados do SCR.data, disponibilizados pelo Banco Central do Brasil.

A proposta consiste em desenvolver um framework capaz de monitorar, explicar e projetar a evolução do risco de crédito em nível de carteira homogênea, incorporando informações macroeconômicas, dinâmica temporal e análises prospectivas compatíveis com os princípios do IFRS 9.

Diferentemente das abordagens tradicionais baseadas em informações individuais das operações de crédito, este estudo concentra-se no comportamento agregado das carteiras, tornando-se adequado para contextos em que não há acesso a dados granulares.

## 2. Contexto de Negócio

A implementação do IFRS 9 introduziu uma mudança estrutural na mensuração das perdas de crédito esperadas (*Expected Credit Loss – ECL*), substituindo o modelo baseado em perdas incorridas por um modelo fundamentado em perdas esperadas.

Um dos principais elementos desse arcabouço é a identificação do aumento significativo do risco de crédito (*Significant Increase in Credit Risk – SICR*), responsável pela migração de exposições do Estágio 1 para o Estágio 2.

Essa migração possui impacto direto sobre:

* O volume de provisões para perdas esperadas;
* A avaliação do risco de crédito das carteiras;
* O planejamento de capital das instituições financeiras;
* A sensibilidade das provisões a cenários macroeconômicos;
* Os processos de monitoramento e governança do risco.

Dessa forma, identificar antecipadamente sinais de deterioração do risco de crédito é um desafio relevante tanto para instituições financeiras quanto para reguladores e pesquisadores.

## 3. Problema de Negócio

Grande parte das implementações do IFRS 9 é baseada em abordagens Bottom-Up, que utilizam informações individuais para estimar parâmetros como:

* Probabilidade de Inadimplência (PD);
* Perda Dada a Inadimplência (LGD);
* Exposição no Momento da Inadimplência (EAD).

Embora essas abordagens sejam amplamente utilizadas, sua aplicação depende da disponibilidade de dados detalhados em nível de contrato ou cliente.

No caso do SCR.data, os dados são disponibilizados de forma agregada, impossibilitando a utilização direta de metodologias tradicionais baseadas em informações individuais.

Diante dessa limitação, surge a seguinte questão:

> É possível identificar e antecipar movimentos de deterioração significativa do risco de crédito utilizando métricas agregadas de carteira, dinâmica temporal e informações macroeconômicas?

## 4. Objetivo de Negócio

Desenvolver uma metodologia Top-Down capaz de:

* Monitorar a deterioração do risco de crédito em nível de carteira;
* Identificar fatores macroeconômicos associados à deterioração do risco;
* Projetar indicadores agregados de deterioração;
* Antecipar eventos de SICR;
* Avaliar a sensibilidade das carteiras a diferentes cenários econômicos.

## 5. Hipóteses de Pesquisa

# Business Understanding

## 1. Project Overview

This project investigates the dynamics of Significant Increase in Credit Risk (SICR) under the IFRS 9 framework using a Top-Down modeling approach applied to aggregated credit portfolio data from the Brazilian Central Bank's SCR.data database.

The study aims to develop a framework capable of monitoring, explaining and forecasting credit risk deterioration at the portfolio level, incorporating macroeconomic information, temporal dynamics and stress testing methodologies.

Unlike traditional IFRS 9 implementations based on loan-level information, this research focuses on portfolio-level behavior, making it suitable for situations where only aggregated information is available.

## 2. Business Context

The implementation of IFRS 9 introduced the Expected Credit Loss (ECL) framework, requiring financial institutions to recognize expected losses using historical, current and forward-looking information.

A key component of this framework is the identification of Significant Increase in Credit Risk (SICR), which determines whether an exposure should migrate from Stage 1 to Stage 2.

Such migration has important consequences:

* Increase in expected credit loss provisions.
* Changes in risk assessment and portfolio monitoring.
* Impact on profitability and capital planning.
* Greater sensitivity to macroeconomic scenarios.

Consequently, institutions need reliable methods to identify and anticipate credit deterioration before default materializes.

## 3. Problem Statement

Most IFRS 9 implementations rely on Bottom-Up methodologies based on individual estimates of:

* Probability of Default (PD)
* Loss Given Default (LGD)
* Exposure at Default (EAD)

Although effective, these approaches require granular data and complex calibration procedures.

Publicly available datasets such as SCR.data do not provide individual-level information, making traditional approaches impractical.

Therefore, the central problem addressed in this project is:

Can aggregate portfolio indicators, combined with macroeconomic information and temporal dynamics, anticipate SICR events consistently with IFRS 9 principles?

## 4. Analytical Objective

Develop a Top-Down framework capable of:

1. Monitoring portfolio-level credit deterioration.
2. Quantifying relationships between macroeconomic conditions and portfolio risk.
3. Forecasting deterioration metrics.
4. Predicting SICR events.
5. Evaluating portfolio sensitivity under alternative economic scenarios.

## 5. Research Hypotheses

### H1 – Macroeconomic Sensitivity

Changes in macroeconomic conditions significantly affect portfolio-level deterioration metrics.

Expected drivers include:

* Interest rates
* Economic activity
* Inflation
* Labor market conditions
* Confidence indicators

### H2 – Predictive Power of Aggregate Indicators

Aggregate deterioration metrics and their temporal dynamics contain predictive information regarding future SICR events.

### H3 – Portfolio Heterogeneity

Different portfolio segments exhibit distinct sensitivities to macroeconomic shocks and stress scenarios.

## 6. Business Questions
